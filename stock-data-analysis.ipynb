### Step 1

```python
#importing the libraries
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import calendar
%matplotlib inline
```

```python
#reading the csv files of (apple, google, amazon) into dataframes 
apple = pd.read_csv('AAPL_data.csv', parse_dates=['date'])
google = pd.read_csv('GOOGL_data.csv', parse_dates=['date'])
amazon = pd.read_csv('AMZN_data.csv', parse_dates=['date'])
```

### Step 2
 

#### in this step i am plotting the open, close and volume of the stocks of apple, amazon and google. It would make more sense and be more clear for comparison in the way I decided to plot them by creating 3 subplots while each is showing the 3 companies for each category.

```python
#creating a list of the columns needed as categories and then looping over the number of plots I need.
#each plot will display the one of the categories for all 3 companies
categories = ['open', 'close', 'volume']

fig = plt.figure(figsize=(25, 10))
for sp in range(0,3):
    ax = fig.add_subplot(1,3,sp+1)
    ax.plot(apple.date, apple[categories[sp]], label='Apple', color='blue')
    ax.plot(google.date, google[categories[sp]], label='Google', color='orange')
    ax.plot(amazon.date, amazon[categories[sp]], label='Amazon', color='magenta')
    plt.title("{} of stocks' yearly variation".format(categories[sp]), weight='bold', fontsize=14)
    plt.xlabel('Year',weight='bold', fontsize=14)
    plt.ylabel("{} of stocks".format(categories[sp]), weight='bold', fontsize=14)
    plt.xticks(weight='bold', fontsize=12, rotation=45)
    plt.yticks(weight='bold', fontsize=12)
    plt.legend()
    plt.grid(color = 'y', linewidth = 0.5)
    

plt.show()
```

### Step 3

#### in the following section, checking the fluctations/seasonality by going through 3 approaches:

#### 1. based on the day of the week

```python
#setting the date as the index for all three dataframes
apple = apple.set_index(apple.date)
google = google.set_index(google.date)
amazon = amazon.set_index(amazon.date)
```

```python
#pulling the weekdays from the date and then creating a box plot for each category and each company
apple['week_days'] = apple['date'].apply(lambda time: time.weekday_name)
for sp in range(0,3):
    plt.figure()
    sns.catplot(x='week_days', y=categories[sp], data=apple, kind='box')
    plt.xticks(weight='bold', fontsize=12, rotation=45)
    plt.yticks(weight='bold', fontsize=12)
    plt.title("Apple {} of stocks' Daily variation".format(categories[sp]), weight='bold', fontsize=14)
    plt.show()
```

```python
google['week_days'] = google['date'].apply(lambda time: time.weekday_name)
for sp in range(0,3):
    plt.figure()
    sns.catplot(x='week_days', y=categories[sp], data=google, kind='box')
    plt.xticks(weight='bold', fontsize=12, rotation=45)
    plt.yticks(weight='bold', fontsize=12)
    plt.title("Google {} of stocks' Daily variation".format(categories[sp]), weight='bold', fontsize=14)
    plt.show()
```

```python
amazon['week_days'] = amazon['date'].apply(lambda time: time.weekday_name)
for sp in range(0,3):
    plt.figure()
    sns.catplot(x='week_days', y=categories[sp], data=amazon, kind='box')
    plt.xticks(weight='bold', fontsize=12, rotation=45)
    plt.yticks(weight='bold', fontsize=12)
    plt.title("Amazon {} of stocks' Daily variation".format(categories[sp]), weight='bold', fontsize=14)
    plt.show()
```

#### 2. based on each month

```python
#pulling the month from the date and then creating a bar chart for each category and each company
apple['month'] = apple['date'].apply(lambda time: time.month)
apple['month'] = apple['month'].apply(lambda x: calendar.month_abbr[x])

for sp in range(0,3):
    plt.figure()
    sns.catplot(x='month', y=categories[sp], data=apple, kind='bar')
    plt.xticks(weight='bold', fontsize=12, rotation=45)
    plt.yticks(weight='bold', fontsize=12)
    plt.title("Apple {} of stocks' monthly variation".format(categories[sp]), weight='bold', fontsize=14)
    plt.show()
```

```python
google['month'] = google['date'].apply(lambda time: time.month)
google['month'] = google['month'].apply(lambda x: calendar.month_abbr[x])

for sp in range(0,3):
    plt.figure()
    sns.catplot(x='month', y=categories[sp], data=google, kind='bar')
    plt.xticks(weight='bold', fontsize=12, rotation=45)
    plt.yticks(weight='bold', fontsize=12)
    plt.title("Google {} of stocks' monthly variation".format(categories[sp]), weight='bold', fontsize=14)
    plt.show()
```

```python
amazon['month'] = amazon['date'].apply(lambda time: time.month)
amazon['month'] = amazon['month'].apply(lambda x: calendar.month_abbr[x])

for sp in range(0,3):
    plt.figure()
    sns.catplot(x='month', y=categories[sp], data=amazon, kind='bar')
    plt.xticks(weight='bold', fontsize=12, rotation=45)
    plt.yticks(weight='bold', fontsize=12)
    plt.title("Amazon {} of stocks' monthly variation".format(categories[sp]), weight='bold', fontsize=14)
    plt.show()
```

#### 3. resampling yearly then plotting high vs low

```python
#resampling
a = apple.resample('Y').mean().ffill()
plt.plot(a.index, a.high, label='high', marker='v')
plt.plot(a.index, a.low, label='low', marker='v')
plt.legend()
plt.title("Apple's stock yearly average variation", weight='bold', fontsize=14)
plt.ylabel('high/low of stocks', weight='bold', fontsize=14)
plt.xlabel('Year', weight='bold', fontsize=14)
plt.xticks(weight='bold', fontsize=12)
plt.yticks(weight='bold', fontsize=12)
plt.show()
```

```python
g = google.resample('Y').mean()
plt.plot(g.index, g.high, label='high', marker='x')
plt.plot(g.index, g.low, label='low', marker='x')
plt.title("Google's stock yearly average variation", weight='bold', fontsize=14)
plt.ylabel('high/low of stocks', weight='bold', fontsize=14)
plt.xlabel('Year', weight='bold', fontsize=14)
plt.xticks(weight='bold', fontsize=12)
plt.yticks(weight='bold', fontsize=12)
plt.legend()
plt.show()
```

```python
am = amazon.resample('Y').mean().ffill()
plt.plot(am.index, am.high, label='high', marker='o')
plt.plot(am.index, am.low, label='low', marker='o')
plt.title("Amazon's stock yearly average variation", weight='bold', fontsize=14)
plt.ylabel('high/low of stocks', weight='bold', fontsize=14)
plt.xlabel('Year', weight='bold', fontsize=14)
plt.xticks(weight='bold', fontsize=12)
plt.yticks(weight='bold', fontsize=12)
plt.legend()
plt.show()
```
